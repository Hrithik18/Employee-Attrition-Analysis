```python
#Load & Inspect Data
import pandas as pd

df = pd.read_csv("../Data/HR_Employee_Attrition.csv")
df.head()
df.info()
```

```python
#checking for null and duplicated values
df.isnull().sum()
df.duplicated().sum()
```

```python
#Attrition Rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
attrition_rate
```

```python
#Department-wise Attrition
pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100
```

```python
#Salary Impact
df.groupby('Attrition')['MonthlyIncome'].mean()
```

```python
#Overtime Impact
pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100
```

```python
#Tenure Analysis
df.groupby('Attrition')['YearsAtCompany'].mean()
```

```python
#Save Cleaned Data
df.to_csv("cleaned_hr_attrition.csv", index=False)
```
